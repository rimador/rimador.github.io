```python
from collections import Counter

def compta_columna(
    input_file,
    output_file,
    columna,        # índex començant a 0
    separador="$"
):
    counter = Counter()

    with open(input_file, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            parts = line.split(separador)
            if len(parts) > columna:
                valor = parts[columna]
                counter[valor] += 1

    ordenat = counter.most_common()

    with open(output_file, "w", encoding="utf-8") as f:
        for valor, count in ordenat:
            f.write(f"{valor}{separador}{count}\n")

    print(f"Fitxer creat: {output_file}")
    print(f"Valors diferents: {len(counter)}")


def comptar_linees(input_path, output_path):
    total_output = 0

    # Sum values from output file
    with open(output_path, "r", encoding="utf-8") as f:
        for line in f:
            parts = line.strip().split("$")
            if len(parts) == 2:
                total_output += int(parts[1])

    # Count lines in input file
    with open(input_path, "r", encoding="utf-8") as f:
        total_input = sum(1 for _ in f)

    print(f"\nTotal al diccionari: {total_input}")
    print(f"Total a l'output: {total_output}")

def compta_parelles(input_file, output_file, col3_index=2, col4_index=3, separador="$"):
    counter = Counter()

    # llegim i comptem parelles
    with open(input_file, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            parts = line.split(separador)
            if len(parts) > max(col3_index, col4_index):
                parella = (parts[col3_index], parts[col4_index])
                counter[parella] += 1

    # ordenar per nombre d'aparicions, de més a menys
    ordenades = counter.most_common()

    # escrivim el resultat a output_file
    with open(output_file, "w", encoding="utf-8") as f:
        for (val3, val4), count in ordenades:
            f.write(f"{val3}{separador}{val4}{separador}{count}\n")

    print(f"Fitxer creat: {output_file}")
    print(f"Parelles úniques trobades: {len(counter)}")


############
funcio = 3
############

if funcio == 1:
    # codi
    output = "codis_ordenats.txt"
    compta_columna(
        input_file="diccionari.5.2.3.txt",
        output_file=output,
        columna=2
    )
    comptar_linees("diccionari.5.2.3.txt", output)

elif funcio == 2:
    # rima consonant
    output = "rimes_consonants_ordenades.txt"
    compta_columna(
        input_file="diccionari.5.2.3.txt",
        output_file=output,
        columna=3
    )
    comptar_linees("diccionari.5.2.3.txt", output)

elif funcio == 3:
    output = "parelles_comptades.txt"
    compta_parelles(
        input_file="diccionari.5.2.3.txt",
        output_file=output,
        col3_index=2,  # columna 3
        col4_index=3   # columna 4
    )

else:
    print("Número no contemplat")



```
